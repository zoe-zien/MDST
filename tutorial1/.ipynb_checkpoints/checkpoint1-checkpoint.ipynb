<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = states_edu[['AVG_MATH_4_SCORE','TOTAL_EXPENDITURE']].dropna()
y = states_edu.loc[X.index]['AVG_READING_4_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
model.score(X_test, y_test)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
Total_Expenditure = 'TOTAL_EXPENDITURE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train['TOTAL_EXPENDITURE'], y_train, color = "lightcoral")
plt.scatter(X_train['TOTAL_EXPENDITURE'], model.predict(X_train), color = "khaki")

plt.legend(['True Training','Predicted Training'])
plt.xlabel('Total Expenditure')
plt.ylabel('Grade 4 Reading Score')
plt.title("Model Behavior On Training Set")
```

```python
 Grade_4_Math_Score = 'AVG_MATH_4_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train['AVG_MATH_4_SCORE'], y_train, color = "lightcoral")
plt.scatter(X_train['AVG_MATH_4_SCORE'], model.predict(X_train), color = "khaki")

plt.legend(['True Training','Predicted Training'])
plt.xlabel('Grade 4 Math Score')
plt.ylabel('Grade 4 Reading Score')
plt.title("Model Behavior On Training Set")
```

<h2> Summary </h2>

Both models accurately depict the behavior shown in the graphs. Interstingly, the predicted training of the second model shows a much more precise linear correlation than the true training. Due to the little correlation between total expenditure and outcome scores, simply increasing eudcation spending may not be a direct solution to improving test scores.
